```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("project_data.csv")
data
```

### 원하는 feature 뽑아오기

```python
def SelectFeature(data, feature_list):
    
    X = data[feature_list].values
    y = data['Class'].values
    
    return X,y;
```

### 테스트 케이스 나누기

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(data, test_size=0.2, random_state=42)

#feature 전체
#feature_list = ['V1','V2','V3','V4','V5','V6','V7','V8','V9','V10','V11','V12','V13','V14','V15',
#               'V16','V17','V18','V19','V20','V21','V22','V23','V24','V25','V26','V27','V28'];

#좋은 feature 선별
feature_list = ['V11','V12','V14','V16','V17'];

train_X,train_y = SelectFeature(train_set,feature_list);
test_X , test_y = SelectFeature(test_set,feature_list);

print(train_X);
print(train_y);
```

### f1 Measure

```python
from sklearn.metrics import f1_score

def CalScore(true,predict):

    f1Score = f1_score(true, predict, average='macro')
    print(f1Score)
    f1Score = f1_score(true, predict, average='micro')
    print(f1Score)
    f1Score = f1_score(true, predict, average='weighted')
    print(f1Score)
    f1Score = f1_score(true, predict, average=None)
    print(f1Score)
```

### Triforce

```python
def Triforce(predict1,predict2,predict3):
    
    tri = (predict1 + predict2 + predict3)/3;
    print(np.sum(tri))
    tri = np.around(tri);
    print(np.sum(tri))
    
    return tri;
    
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

logit = LogisticRegression(C = 10000)
logit.fit(train_X, train_y)

#predict = logit.predict(train_X)
#CalScore(train_y,predict)

predict1 = logit.predict(test_X)
CalScore(test_y,predict1)
```

### SVM

```python
from sklearn.svm import SVC

svm = SVC(kernel='rbf')
svm.fit(train_X, train_y)

#predict = svm.predict(train_X)
#CalScore(train_y,predict)

predict2 = svm.predict(test_X)
CalScore(test_y,predict2)
```

### Decision Tree Classifier

```python
from sklearn import tree

decisionTree = tree.DecisionTreeClassifier()
decisionTree.fit(train_X, train_y)


predict = decisionTree.predict(train_X)
CalScore(train_y,predict)

predict = decisionTree.predict(test_X)
CalScore(test_y,predict)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification

rfc = RandomForestClassifier(n_estimators=100, max_depth=2,random_state=0)
rfc.fit(train_X, train_y)

#predict = rfc.predict(train_X)
#CalScore(train_y,predict)

predict3 = rfc.predict(test_X)
CalScore(test_y,predict3)
```

```python
triforce = Triforce(predict1,predict2,predict3)
print(triforce)

CalScore(test_y,triforce)
```

### Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier

gbc = GradientBoostingClassifier()
gbc.fit(train_X, train_y)

predict = gbc.predict(test_X)
CalScore(test_y,predict)

```

```python
import xgboost as xgb

xgb_model = xgb.XGBClassifier()
xgb_model.fit(train_X, train_y)

predict = xgb_model.predict(test_X)
CalScore(test_y,predict)
```

### MLP Classifier

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)  
mlp.fit(train_X, train_y)  

predict = mlp.predict(test_X)
CalScore(test_y,predict)
```

### k-NN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(train_X, train_y)

predict = knn.predict(test_X)
CalScore(test_y,predict)
```

### svm 파라미터 튜닝

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.001,0.01,0.1, 1, 10, 100,1000], 'gamma': [10, 1, 0.1, 0.01, 0.001, 0.00001]}


clf = GridSearchCV(SVC(), param_grid, verbose=1)
clf.fit(train_X, train_y)

print("Best Parameters:\n", clf.best_params_)
print("Best Estimators:\n", clf.best_estimator_)
```
